```python
df11=pd.DataFrame({'Date':['2023-01-01','2023-01-02','2023-01-03','2023-01-04','2023-01-05']})
df11
```

```python
from datetime import datetime
```

```python
d=list(df11['Date'])
d1=[]
for i in d:
    d1.append(datetime.strptime(i,'%Y-%m-%d').strftime('%A'))
d1
```

```python
df11['Weekday']=d1
```

```python
df11
```
